## Explanation
1. Not all launch sites are very close to the Equator line but they are in relatively low latitudes compared to other global locations.
- The Equator is at 0° latitude. 
- The Florida launch sites (CCAFS & KSC) are closer to the Equator (~28° N) while VAFB SLC-4E (California) is further north (~34° N) meaning it is less optimal for equatorial launches but useful for polar orbits.
- Launching closer to the Equator takes advantage of Earth’s rotational speed (~1670 km/h), which helps rockets save fuel when launching eastward.

2. All launch sites are near coastal areas.
This is likely intentional for:
- Safety reasons: Rockets may fail during launch, and launching over the ocean reduces the risk to populated areas.
- Efficient Launch Trajectories: Eastward launches (for geostationary & low Earth orbits) take advantage of Earth's rotation.Also, polar orbits (often from VAFB SLC-4E) require launching southward over the Pacific Ocean.
- Logistics and infrastructure: Rockets & boosters are often transported by sea (e.g., SpaceX droneships for landings).

## TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

## Findings
Are launch sites in close proximity to railways?
Yes, most launch sites are located relatively close to railways, but not directly next to them.

Explanation:
Rockets and fuel need large-scale transportation, so railways nearby make sense.
However, launch sites require open land, so they are not built directly next to railways.

Are launch sites in close proximity to highways?
Yes, launch sites are usually near highways for logistical support.

Explanation:
Rockets, payloads, and fuel need heavy transport trucks. Quick emergency access is essential.
But not too close, as shockwaves from launches could damage infrastructure.

Are launch sites in close proximity to coastline?
Yes, most launch sites are built very close to the coastline.

Explanation:
Safety Reason: Rockets are launched over the ocean to avoid damage to populated areas if failures occur.
Lower Risk: Debris falls into the ocean instead of land.
Open Space: No tall buildings or mountains to obstruct launches.

Do launch sites keep certain distance away from cities?
Yes, launch sites are far from major cities.

Explanation:
Rocket Launch Shockwaves: The launch produces powerful sound waves and vibrations, which could damage buildings in urban areas.
Public Safety: In case of launch failure, large urban populations are not at risk.
Airspace Control: Less air traffic interference.
